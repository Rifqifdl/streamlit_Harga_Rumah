# Jumlah rumah berdasarkan tahun pembangungan

```python
made = df.groupby('made').count()[['price']].sort_values(by='price',ascending=True).reset_index()
made = made.rename(columns={'price':'numberOfHouse'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=made['made'], y=made['numberOfHouse'], color='skyblue')
plt.xticks(rotation=60)
```

# Berdasarkan Jumlah pemilik yang pernah tinggal

```python
room = df.groupby('numPrevOwners').count()[['price']].sort_values(by='price',ascending=True).reset_index()
room = room.rename(columns={'price':'numberOfHouse'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=room['numPrevOwners'], y=room['numberOfHouse'], color='skyblue')
plt.xticks(rotation=60)
```

# Distribusi luas lahan

```python
plt.figure(figsize=(15,5))
sns.distplot(df['price'])
```

# Seleksi fitur

```python
features = ['squareMeters', 'numberOfRooms', 'floors', 'numPrevOwners', 'basement']
x = df[features]
y = df['price']
x.shape, y.shape
```

# Split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test, y_test)
print('Akurasi:', score)
```

# Membuat inputan model Regresi Linier

```python
#squaremeters #numberofrooms #floors #numprevowners #hasyard #haspool #basement
input_data = np.array([[20000,10,4,2,320]])

prediction = lr.predict(input_data)
print('Estimasi Harga: ', prediction)
```
